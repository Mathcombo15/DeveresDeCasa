1- Entrada do usuário:


```python
# Solicitar ao usuário que insira uma frase.
# Verificar se a entrada não está vazia e tratar o erro caso necessário. (crie uma nova função no bloco de funções)

strEntradaUsuario = input("Sr. Usuário, insira uma frase: ")
if strEntradaUsuario == ("") :
    print("error: campo deve ser preenchido")
 else:
    print(strEntradaUsuario)
```

2- Análise da frase:

```python
# Contagem de Caracteres: Calcular o número total de caracteres da frase (incluindo espaços).
# Contagem de Palavras: Dividir a frase em palavras (utilizando o método `split()`) e determinar quantas palavras existem.
# Maior Palavra: Identificar a palavra com o maior número de caracteres dentre as encontradas.

strFraseAnalise = input("Digite uma frase: ")
if strFraseAnalise == ("") :
    print("error: campo deve ser preenchido")
else:
    print('A frase digitada tem: ' + str(len(strFraseAnalise)) +' caracteres') # contagem de caracteres
    print('A frase digitade tem: ' + str(len(strFraseAnalise.split())) +' palavras > ') # contagem de palavras
    print(strFraseAnalise.split())
    print(max(strFraseAnalise.split())) # identificação da maior palavra
```

3- Manipulação e Formatação:

```python
# Inversão da Frase:
# Inverter a frase por meio dos caracteres (utilizando slicing com `[::-1]`);
# Inverter a ordem das palavras e reconstruir a frase.
# Alteração de Caixa: Converter a frase para todas as letras maiúsculas e para todas as letras minúsculas.
# Tupla de Palavras: Criar uma tupla contendo todas as palavras da frase.

strManipulacaoeFormatacao = input("Digite uma frase: ")
if strManipulacaoeFormatacao == ("") :
    print("error: campo deve ser preenchido")
else:
  print('Frase invertida: ' + str(strManipulacaoeFormatacao[::-1]))
  print('Frase invertida por meio dos caracteres: ' +str(strManipulacaoeFormatacao.split()[::-1]))
  print(print('Frase invertida para o estado inicial: ' + str(strManipulacaoeFormatacao[::+1])))
  print('Frase em letra maiúscula: ' + strManipulacaoeFormatacao.upper())
  print('Tupla composta pela frase inserida: '+ str(tuple((strManipulacaoeFormatacao.split()))))
```

4- Saída Formatada:

```python
# Utilizar f-strings para exibir os resultados de forma organizada, apresentando as seguintes informações:
# Número de caracteres da frase.
# Número de palavras.
# A palavra com maior comprimento.
# A frase invertida (por caracteres e por palavras).
# A frase em maiúsculas e minúsculas.
# A tupla formada pelas palavras da frase.

strSaidaFormatada = input("Digite uma frase: ")
NumCaracteres = len(strSaidaFormatada)
NumPalavras = len(strSaidaFormatada.split())
MaiorPalavra = max(strSaidaFormatada.split())
FraseIvertida = str(strSaidaFormatada[::-1])
FraseIvertidaPalavras = str(strSaidaFormatada.split()[::-1])
FraseInvertidaInicial = str(strSaidaFormatada[::+1])
FraseMaius = strSaidaFormatada.upper()
FraseMinus = strSaidaFormatada.lower()
TuplaPalavras = tuple((strSaidaFormatada.split()))

if strSaidaFormatada == ("") :
    print("error: campo deve ser preenchido")
else:
  print(f'''A frase insirida pelo usuário tem {NumCaracteres} caracteres, composta por {NumPalavras} palavras,
  sendo {MaiorPalavra} a maior palavra desta frase. A frase com seus caracteres invertidos fica da seguinte forma: {FraseIvertida},
   e fazendo a iversão das palavras fica assim: {FraseIvertidaPalavras}. Outra forma de inversão é: {FraseInvertidaInicial}.
   Maiúscula e Minúscula respectivamente a frase fica: {FraseMaius},{FraseMinus} e por fim temos a tupla da frase digitada:{TuplaPalavras} ''')
```
